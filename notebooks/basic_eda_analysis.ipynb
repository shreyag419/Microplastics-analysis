```
import pandas as pd
import matplotlib.pyplot as plt
```

```
delhi = pd.read_csv("Delhi_AQI_Dataset.csv")
mumbai = pd.read_csv("Mumbai_AQI_Dataset.csv")
soil = pd.read_csv("soil_pollution_diseases.csv")
marine = pd.read_csv("Marine_Microplastics.csv")
```

```
print("Delhi Info:\n", delhi.info())
print("Mumbai Info:\n", mumbai.info())
print("Soil Info:\n", soil.info())
print("Marine Info:\n", marine.info())
```

```
print("\nDelhi Stats:\n", delhi.describe())
print("\nMumbai Stats:\n", mumbai.describe())
print("\nSoil Stats:\n", soil.describe())
print("\nMarine Stats:\n", marine.describe())
```

```
print(delhi.isnull().sum())
print(mumbai.isnull().sum())
print(soil.isnull().sum())
print(marine.isnull().sum())
```

```
print("Delhi AQI")
plt.figure()
delhi['AQI'].hist()
plt.title("Delhi AQI Distribution")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()
```

```
plt.figure()
mumbai['AQI'].hist()
plt.title("Mumbai AQI Distribution")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()
```

```
print("Soil Pollution")
plt.figure()
soil['Pollutant_Concentration_mg_kg'].hist()
plt.title("Soil Pollution Distribution")
plt.xlabel("Concentration")
plt.ylabel("Frequency")
plt.show()
```

```
print("Marine")
plt.figure()
marine['Measurement'].hist()
plt.title("Marine Microplastics Distribution")
plt.xlabel("Measurement")
plt.ylabel("Frequency")
plt.show()
```

```
print("Air Pollution Comparison")
delhi_avg = delhi[['PM2.5','PM10','NO2','SO2','CO','O3']].mean()
mumbai_avg = mumbai[['PM2.5','PM10','NO2','SO2','CO','O3']].mean()

combined = pd.DataFrame({
    "Delhi": delhi_avg,
    "Mumbai": mumbai_avg
})

combined.plot(kind='bar')
plt.title("Delhi vs Mumbai Pollution")
plt.xlabel("Pollutants")
plt.ylabel("Average Values")
plt.xticks(rotation=45)
plt.show()
```

```
print("Soil Disease Count")
soil['Disease_Type'].value_counts().plot(kind='bar')
plt.title("Disease Distribution (Soil)")
plt.show()
```

```
print("Marine Ocean Distribution")
marine['Oceans'].value_counts().plot(kind='bar')
plt.title("Ocean Distribution")
plt.show()
```

```
print("AQI Trend")
plt.figure()
plt.plot(delhi['Date'], delhi['AQI'])
plt.title("Delhi AQI Trend")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.show()
```

```
plt.figure()
plt.plot(mumbai['Date'], mumbai['AQI'])
plt.title("Mumbai AQI Trend")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.show()
```

```
print("Soil over time")
soil['Date_Reported'] = pd.to_datetime(soil['Date_Reported'])
soil_sorted = soil.sort_values('Date_Reported')

plt.figure()
plt.plot(soil_sorted['Date_Reported'], soil_sorted['Pollutant_Concentration_mg_kg'])
plt.title("Soil Pollution Over Time")
plt.show()
```

```
print("Marine over time")
marine['Date'] = pd.to_datetime(marine['Date'], format='%d-%m-%Y', errors='coerce')
marine_sorted = marine.sort_values('Date')

plt.figure()
plt.plot(marine_sorted['Date'], marine_sorted['Measurement'])
plt.title("Marine Pollution Over Time")
plt.show()
```

```
print("Delhi correlation")
corr_delhi = delhi[['AQI','PM2.5','PM10','NO2','SO2','CO','O3']].corr()
print("\nDelhi Correlation:\n", corr_delhi)

print("Soil correlation")
corr_soil = soil.select_dtypes(include='number').corr()
print("\nSoil Correlation:\n", corr_soil)
```

```
print("AQI boxplot")
plt.figure()
delhi['AQI'].plot.box()
plt.title("Delhi AQI Outliers")
plt.show()

plt.figure()
marine['Measurement'].plot.box()
plt.title("Marine Outliers")
plt.show()
```
